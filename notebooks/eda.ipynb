```python
import seaborn as sns
import matplotlib.pyplot as plt

categ_variable = [col for col in numeric_df.columns if df_clean[col].nunique() <= 4]
categ_variable.remove('fraud_reported')

for categ_var in categ_variable:
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(data=df_clean, x=categ_var, hue='fraud_reported')

    cross_tab = df_clean.groupby([categ_var, 'fraud_reported']).size().reset_index(name='count')
    total_per_group = df_clean.groupby(categ_var).size().to_dict()

    for p in ax.patches:
        count = int(p.get_height())
        x_pos = p.get_x() + p.get_width() / 2.
        xticks = ax.get_xticks()
        tick_labels = [t.get_text() for t in ax.get_xticklabels()]
        idx = min(range(len(xticks)), key=lambda i: abs(xticks[i] - x_pos))
        category_label = tick_labels[idx]
        total = total_per_group.get(category_label, 0)
        pct = f"{(count / total * 100):.1f}%" if total > 0 else ""

        ax.annotate(f"{count}\n{pct}", 
                    (x_pos, p.get_height()), 
                    ha='center', va='bottom', fontsize=9)

    plt.title(f"Fraud Count by {categ_var}")
    plt.xlabel(categ_var)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
categ = ['auto_make', 'policy_csl', 'insured_occupation', 'age', 'auto_model']

for col in categ:
    print(f"\n--- Fraud rate by {col} ---\n")
    
    # Moyenne de fraude par modalité
    fraud_rate = df_clean.groupby(col)['fraud_reported'].mean()
    
    # Nombre d’occurrences par modalité
    counts = df_clean[col].value_counts()
    
    # Jointure des deux infos dans un DataFrame
    summary = pd.concat([fraud_rate, counts], axis=1)
    summary.columns = ['fraud_rate', 'count']
    
    # Trié par taux de fraude décroissant
    print(summary.sort_values(by='fraud_rate', ascending=False))
```

```python
categ = ['auto_make', 'policy_csl', 'insured_occupation', 'age', 'auto_model', 'collision_type', 'incident_type',
        'insured_occupation' , 'incident_hour_of_the_day', 'collision_type', 'incident_severity',
         'authorities_contacted', 'insured_relationship', 'property_damage']

for col in categ:
    print(f"--- Fraud rate by {col} ---")
    rates = df_clean.groupby(col)['fraud_reported'].mean().sort_values(ascending=False)
    counts = df_clean[col].value_counts()
    
    summary = pd.concat([rates, counts], axis=1)
    print(summary)
    print('\n')
```

## 6. Key Observations

- The presence of a police report is strongly associated with a lower fraud rate.
- Most frauds occur when 1 or 3 vehicles are involved.
- Some financial fields like `vehicle_claim` and `property_claim` show moderate correlation with `fraud_reported`.
- No major impact of `policy_state` or any others variables.
- Individuals working in **executive, farming, or repair-related occupations** show significantly higher fraud rates compared to other professions. but risk of correlation between `insured_occupation` variable and the `advanced_study`.
- Certain car models such as **BMW X6, Mercedes ML350, and Chevrolet Silverado** are more likely to be involved in fraudulent claims, with fraud rates above 40%.
- Fraud rates also vary across age, with spikes observed in small age groups (e.g., 19, 22, 51, 52), which should be interpreted carefully due to low sample sizes.

- These observations support the creation of risk-based features, which may enhance the predictive performance of the model by grouping high-risk behaviors or profiles together.
