## 4. Correlation with Fraud

```python
df_clean.select_dtypes(include='number').corr()
```

```python
# Correlation between vehicule_claim amount and the fraud_reported

numeric_df = df_clean.select_dtypes(include='number')
non_binary_cols = [col for col in numeric_df.columns if df_clean[col].nunique() > 2] #to exclude the dummies numeric variables

if 'fraud_reported' not in non_binary_cols:
    non_binary_cols.append('fraud_reported')
    
corr = df_clean[non_binary_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr[['fraud_reported']].sort_values(by='fraud_reported', ascending=False), 
            annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation with Fraud Reported")
plt.show()
```

## 5. Fraud by Categorical Variables

```python
# Less fraud in the cases reported by the Police
sns.countplot(data=df_clean, x='police_report_available', hue='fraud_reported')
plt.title("Fraud by Police Report Availability")
plt.xlabel("Police Report")
plt.ylabel("Count")
plt.legend(title="Fraud")
plt.show()
```

```python
categ_variable = [col for col in numeric_df.columns if df_clean[col].nunique() <= 4]
categ_variable.remove('fraud_reported')

for categ_var in categ_variable:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df_clean, x=categ_var, hue='fraud_reported')
    plt.xticks(rotation=45)
    plt.title(f"Fraud Count by {categ_var}")
    plt.show()
    
# Much less fraud where there police report is available (whitout taking in account the Unkown values)
# Concerned most of the time 1 or 3 vehicule involved
```

## 6. Key Observations

- The presence of a police report is strongly associated with a lower fraud rate.
- Most frauds occur when 1 or 3 vehicles are involved.
- Some financial fields like `vehicle_claim` and `property_claim` show moderate correlation with `fraud_reported`.
